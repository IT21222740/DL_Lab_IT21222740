```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import io



```

```python
file_name = list(uploaded.keys())[0]

# Load the dataset
df = pd.read_csv(file_name)

#Create a new dataframe by filtering the specified columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

print(new_df.head(20))
mean_values = new_df.mean()
std_values = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean Values:\n", mean_values)
print("\nStandard Deviation:\n", std_values)
print("\nQuartile Values:\n", quartiles)

```

```python

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

for i, feature in enumerate(features):
    sns.histplot(new_df[feature], kde=True, ax=axes[i])
    axes[i].set_title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()

#Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()


g = sns.PairGrid(new_df)


g.map_diag(sns.histplot, kde=True)

# Plots on the lower half
g.map_lower(sns.kdeplot, cmap="Blues_d")

# Plots on the upper half
g.map_upper(sns.scatterplot)

plt.show()
```
