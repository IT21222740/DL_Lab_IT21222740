```python
from __future__ import division, print_function, unicode_literals
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_array_large = np.random.exponential(scale=1.0, size=(100000, 1))
print(exp_array_large)
```

```python
uniform_array = np.random.uniform(low=0.0, high=1.0, size=(100000, 1))
normal_array = np.random.normal(loc=0.0, scale=1.0, size=(100000, 1))
```

```python
plt.figure(figsize=(14, 8))

# Histogram for exponential distribution
plt.hist(exp_array_large, bins=100, alpha=0.5, label='Exponential', color='blue', density=True)

# Histogram for uniform distribution
plt.hist(uniform_array, bins=100, alpha=0.5, label='Uniform', color='green', density=True)

# Histogram for normal distribution
plt.hist(normal_array, bins=100, alpha=0.5, label='Normal', color='red', density=True)

plt.axis([-2.5, 2.5, 0, 1.1])
plt.title("Random distributions")
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate Z
z = x**2 + y**2

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$Z = X^2 + Y^2$')

plt.show()
```

```python
# Pandas for managing datasets
import pandas as pd
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df
```

```python
# Select the relevant columns
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features].dropna()  # Drop rows with any NaN values

# Calculate Pearson correlation coefficient
pearson_corr = data.corr(method='pearson')

# Calculate Spearman rank correlation coefficient
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

# Pearson heatmap
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=ax[0])
ax[0].set_title('Pearson Correlation')

# Spearman heatmap
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=ax[1])
ax[1].set_title('Spearman Correlation')

plt.show()
```
